```python
#Import the libraries
import pandas as pd
import numpy as np
```

```python
#Read the Data Set from CSV File
df = pd.read_csv('CountryData.csv')
df.head()
```

# Finding the Missing Values

```python
#Lets try to find if values are missing from the dataset

df.isnull().tail()
```

```python
#Lets use the function notnull() to check which values are not null

df.notnull().tail()
```

```python
# Let's do another operation to find number of missing values in each columns

df.isnull().sum()
```

# Removing Complete Null Rows

```python
#Now we want to find that which rows are complete null and remove them
data = df
null_rows = data.index[data.isnull().all(1)]
null_rows
```

```python
df.shape
```

```python
#Drop all the Rows that are complete Null

df.drop(null_rows, inplace = True)
```

```python
df.shape
```

```python
# now we want to drop all the values which are missing in our data set
# Drop the row if any of them have 'NaN' Values

df.dropna(how = 'any', inplace = True)
```

```python
# Lets check that how many rows and Columns we have after cleaning the data
df.shape
```

```python
# Again, ensure that If the dataset contain any missing values
# We can see that there are no Columns that contain any missing values

df.isnull().sum()
```
